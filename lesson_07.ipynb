```python
# Zadanie 9

wagi = [70, 85, 62, 95, 78]
wzrosty = [1.75, 1.80, 1.65, 1.78, 1.72]

def bmi(wagi_data, wzrosty_data):
    
    for v in range(len(wagi_data)):
        bmi_value = wagi_data[v] / (wzrosty_data[v] ** 2)

        if bmi_value < 18.5: opis = 'niedowaga'
        elif 18.5 <= bmi_value < 25: opis = 'norma'
        else: opis = 'nadwaga'

        print(f'Osoba {v+1}: BMI = {bmi_value:.2f} {opis}')
        
bmi(wagi, wzrosty)
```

```python
# Zadanie 9 (wersja zip)

wagi = [70, 85, 62, 95, 78]
wzrosty = [1.75, 1.80, 1.65, 1.78, 1.72, 1]

def bmi(wagi_data, wzrosty_data):

    if len(wagi) == len(wzrosty):

        for v, i in zip(wagi_data, wzrosty_data):
            bmi_value = v / (i ** 2)

            if bmi_value < 18.5: opis = 'niedowaga'
            elif 18.5 <= bmi_value < 25: opis = 'norma'
            else: opis = 'nadwaga'

            print(f'Osoba {v+1}: BMI = {bmi_value:.2f} {opis}')
    else: 
        print('Listy są niekompatybilne')

        
bmi(wagi, wzrosty)
```

```python
# Zadanie 10

import math


n = list(range(2, 101))
b = list(range(5, 55))

def primal_numbers(lista):

    primals = []

    for i in lista:

        if i < 2: continue

        czy_pierwsza = True
        
        for dzielnik in range(2, int(math.sqrt(i)) + 1):
            if i % dzielnik == 0:
                czy_pierwsza = False
                break
        
        if czy_pierwsza:
            primals.append(i)

    return primals

print(f'Liczy pierwsze: {primal_numbers(n)}')
print(f'Liczba znalezionych: {len(primal_numbers(n))}')
print(f'Liczy pierwsze: {primal_numbers(b)}')
print(f'Liczba znalezionych: {len(primal_numbers(b))}')
```

```python
# Zadanie 11

import calendar
import locale
import matplotlib.pyplot as plt


locale.setlocale(locale.LC_TIME, 'Polish')

sells = [1200, 1350, 1290, 1480, 1520, 1510, 1680, 1720, 1700, 1890, 1950, 2100]

for i in range(1, len(sells)):

    stary_miesiac = sells[i-1]
    nowy_miesiac = sells[i]

    if nowy_miesiac > stary_miesiac:

        wzrost_procentowy = ((nowy_miesiac - stary_miesiac) / stary_miesiac) * 100
        wzrost_kwota = nowy_miesiac - stary_miesiac

        print(f'{calendar.month_name[i+1].capitalize():12} | {wzrost_kwota:4} PLN | {wzrost_procentowy:.2f} %')

# 1. Tworzenie wykresu
miesiace = [calendar.month_name[i].capitalize() for i in range(1, 13)]

plt.figure(figsize=(12, 6))

# 2. Rysowanie słupków
plt.bar(miesiace, sells, color='skyblue', edgecolor='navy') # Wykres słupkowy

# 3. Dodawanie opisów i tytułów
plt.title('Sprzedaż w poszczególnych miesiącach', fontsize=14)
plt.xlabel('Miesiąc', fontsize=12)
plt.ylabel('Kwota sprzedaży (zł)', fontsize=12)

# 4. Opcjonalne: Dodanie siatki pomocniczej
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 5. Wyświetlenie wykresu
plt.show()
```
