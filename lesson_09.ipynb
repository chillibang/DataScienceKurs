```python
# Zadanie 9

class FiltrowanieDanych:
    
    def __init__(self):
        self.nazwa = "Usuwanie ujemnych wartości"

    def wykonaj(self, dane):
        print(f"Wykonuję: {self.nazwa}")
        return [x for x in dane if x >= 0]

class Zaokraglanie:
    
    def __init__(self):
        self.nazwa = "Zaokrąglanie wartości"
    
    def wykonaj(self, dane):
        print(f"Wykonuję: {self.nazwa}")
        return [round(x) for x in dane]
    
class Skalowanie:
    
    def __init__(self):
        self.nazwa = "Skalowanie wartości"

    def wykonaj(self, dane):
        print(f"Wykonuję: {self.nazwa}")
        return [x * 2 for x in dane]
    
class Sumowanie:
    
    def __init__(self):
        self.nazwa = "Sumowanie wartości"

    def wykonaj(self, dane):
        print(f"Wykonuję: {self.nazwa}")
        return sum(dane)

class Pipeline:
    
    def __init__(self, nazwa):
        self.nazwa = nazwa
        self.kroki = []
    
    def dodaj_krok(self, krok):
        self.kroki.append(krok)
    
    def wykonaj(self, dane):
        print(f"Uruchamiam pipeline: {self.nazwa}")
        for krok in self.kroki:
            dane = krok.wykonaj(dane)
        return dane

pipeline = Pipeline("Przetwarzanie danych")
pipeline.dodaj_krok(FiltrowanieDanych())
pipeline.dodaj_krok(Zaokraglanie())
pipeline.dodaj_krok(Skalowanie())
pipeline.dodaj_krok(Sumowanie())



dane = [1.7, -3.2, 5.8, -1.1, 8.3, 2.4]
wynik = pipeline.wykonaj(dane)
print(f"Wynik końcowy: {wynik}")

```

```python
# Zadanie 10

class Produkt:
    def __init__(self, nazwa, cena):
        self.nazwa = nazwa
        self.cena = cena
    
    def __str__(self):
        return f"Produkt: {self.nazwa}, Cena: {self.cena}"
    
class Zamówienie:
    def __init__(self):
        self.produkty = []
    
    def dodaj_produkt(self, produkt):
        self.produkty.append(produkt)
    
    def oblicz_cene(self):
        return sum(produkt.cena for produkt in self.produkty)

produkt1 = Produkt("Laptop", 3000)
produkt2 = Produkt("Smartfon", 1500)
produkt3 = Produkt("Tablet", 2000)
produkt4 = Produkt("Laptop", 3000)
produkt5 = Produkt("Smartfon", 1500)

zamówienie = Zamówienie()
zamówienie.dodaj_produkt(produkt1)
zamówienie.dodaj_produkt(produkt2)
zamówienie.dodaj_produkt(produkt3)
zamówienie.dodaj_produkt(produkt4)
zamówienie.dodaj_produkt(produkt5)

cena_zamówienia = zamówienie.oblicz_cene()
print(f"Cena zamówienia: {cena_zamówienia}")    
```

```python
# Zadanie 12

class Figura:
    def __init__(self, nazwa):
        self.nazwa = nazwa
    
    def oblicz_pole(self):
        pass

class Prostokat(Figura):
    def __init__(self, szerokosc, wysokosc):
        super().__init__("Prostokąt")
        self.szerokosc = szerokosc
        self.wysokosc = wysokosc
    
    def oblicz_pole(self):
        return self.szerokosc * self.wysokosc   

class Kolo(Figura):
    def __init__(self, promien):
        super().__init__("Koło")
        self.promien = promien
    
    def oblicz_pole(self):
        import math
        return math.pi * self.promien ** 2

prostokat = Prostokat(4, 5)
kolo = Kolo(3)

print(f"Pole {prostokat.nazwa}: {prostokat.oblicz_pole()}")
print(f"Pole {kolo.nazwa}: {kolo.oblicz_pole():.2f}")
print(f"Suma pól: {prostokat.oblicz_pole() + kolo.oblicz_pole():.2f}")

```

```python
class Figura:
    def __init__(self, nazwa):
        self.nazwa = nazwa
    
    def oblicz_pole(self, **parametry):
        if self.nazwa == "Prostokąt":
            return parametry['szerokosc'] * parametry['wysokosc']   
        elif self.nazwa == "Koło":
            import math
            return parametry['promien'] ** 2 * math.pi

class Prostokat(Figura):
    def __init__(self, szerokosc, wysokosc):
        super().__init__("Prostokąt")
        self.szerokosc = szerokosc
        self.wysokosc = wysokosc

class Kolo(Figura):
    def __init__(self, promien):
        super().__init__("Koło")
        self.promien = promien

prostokat = Prostokat(4, 5)
kolo = Kolo(3)

print(f"Pole {prostokat.nazwa}: {prostokat.oblicz_pole(szerokosc=prostokat.szerokosc, wysokosc=prostokat.wysokosc)}")
print(f"Pole {kolo.nazwa}: {kolo.oblicz_pole(promien=kolo.promien):.2f}")
```
