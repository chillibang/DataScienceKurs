```python
# Zadanie 9

from collections import defaultdict


products = {'laptop': 'elektronika',
            'klawiatura': 'elektronika',
            'krzesło': 'meble',
            'biurko': 'meble',
            }

category = defaultdict(list)
for product, cat in products.items():
    category[cat].append(product)   

result = dict(category)
print(result)

```

```python
# Zadanie 12

transakcje = [
    ('2024-01-15', 'Elektronika', 1200),
    ('2024-01-15', 'Meble', 450),
    ('2024-01-16', 'Elektronika', 800),
    ('2024-01-16', 'Elektronika', 350),
    ('2024-01-15', 'Odzież', 200),
    ('2024-01-16', 'Meble', 600)
]

raport_sell = {}

for data, kategoria, kwota in transakcje:
    if data not in raport_sell:
        raport_sell[data] = {}
    
    if kategoria not in raport_sell[data]:
        raport_sell[data][kategoria] = 0
    
    raport_sell[data][kategoria] += kwota

print(raport_sell)
```
