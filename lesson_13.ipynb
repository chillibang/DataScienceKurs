```python
# Zadanie 10

"""
Używając danych sprzedażowych, pogrupuj wg kategorii.
Użyj HAVING aby znaleźć tylko kategorie z przychodem > 1000 PLN
Oblicz liczbę transakcji, łączny przychód i średnią dla każdej kategorii.
Posortuj wg rocznego przychodu.
"""

import pandas as pd
import sqlite3


data = {
    'id': [1, 2, 3, 4, 5, 6],
    'kategoria': ['Elektronika', 'Dom', 'Elektronika', 'Moda', 'Dom', 'Elektronika'],
    'przychod': [800, 150, 400, 2000, 900, 50]
}
df = pd.DataFrame(data)

conn = sqlite3.connect(':memory:')
df.to_sql('sprzedaz', conn, index=False)

query = """
SELECT 
    kategoria,
    COUNT(id) AS liczba_transakcji,
    SUM(przychod) AS laczny_przychod,
    AVG(przychod) AS sredni_przychod
FROM sprzedaz
GROUP BY kategoria
HAVING SUM(przychod) > 1000
ORDER BY laczny_przychod DESC
"""

result_sql = pd.read_sql_query(query, conn)
print("Wynik SQL (HAVING > 1000):")
print(result_sql)
```
