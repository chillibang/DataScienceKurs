```python
# Zadanie 9

value = [75, 100, 50, 62.5]
col_min = 50
col_max = 100

for i in value:
    normalized_value = (i - col_min) / (col_max - col_min)

    if normalized_value >= 0 and normalized_value <= 1:
        norm_bool = True
    else:
        norm_bool = False

    print(f"{'Original value':.<20} {i:<5} {'Normalized value':.<20} {normalized_value:<4} {'Is normalized':.<20} {norm_bool}")
```
