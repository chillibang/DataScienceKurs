```python
# Zadanie 9

def normalize_minmax(dane):
    if not dane:
        return []
    
    val_min = min(dane)
    val_max = max(dane)
    
    if val_max == val_min:
        return [0.0] * len(dane)
    
    znormalizowane = []
    for x in dane:
        wynik = (x - val_min) / (val_max - val_min)
        znormalizowane.append(round(wynik, 3))
        
    return znormalizowane

wynagrodzenia = [3500, 4200, 5800, 7200, 9500, 4800, 6100]
wynik = normalize_minmax(wynagrodzenia)

print(f"Oryginalne dane: {wynagrodzenia}")
print(f"Znormalizowane:  {wynik}")
```
