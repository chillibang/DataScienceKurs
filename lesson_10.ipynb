```python
# Zadanie 10

"""Dana tablica 5 studentów x 4 egzaminów. 
Dodaj bonus do każdego egzaminu [5, 10, 5, 10]. 
Pomnóż wyniki każdego studenta przez mnożnik [1.0, 1.1, 0.95, 1.15, 1.0]
Użyj broadcastingu (bez pętli), wypisz wyniki przed i po"""

import numpy as np

exams = np.array([
    [78, 85, 92, 88],
    [65, 72, 68, 70],
    [90, 88, 95, 92],
    [55, 60, 58, 62],
    [82, 79, 85, 80]
])
print(f"* Lista studentów i ich wyników z egzaminów:\n{exams}\n")
print(f"* Wielkość tablicy studentów: {exams.shape}\n")

bonus = np.array([5, 10, 5, 10])
print(f"* Bonusy za egzaminy:\n{bonus}\n")
print(f"* Wielkość tablicy bonus: {bonus.shape}\n")

exams_plus_bonus = exams + bonus
print(f"* Wyniki po dodaniu bonusów:\n{exams_plus_bonus}\n")

multipliers = np.array([1.0, 1.1, 0.95, 1.15, 1.0]).reshape(5, 1)
exams_final = exams_plus_bonus * multipliers
print(f"* Mnożniki:\n{multipliers}\n")
print(f"* Wyniki ostateczne:\n{exams_final}\n") 
```

```python
# Zadanie 11

"""Sprzedaż 6 produktów w 30 dni
Oblicz trend sprzedaży per produkt (różnica między ostatnimi 7 dniami a pierwszymi 7 dniami)
Oblicz najlepszy tydzień (dni 0-6, 7-13, 14-20, 21-27) dla całej sprzedaży (suma wszystkich produktów) 
Wykres: sprzedaż per produkt w czasie (linie)
Użyj slicingu i agregacji, bez pętli
Wizualizacja: matplotlib"""

import numpy as np
import matplotlib.pyplot as plt

sales = np.random.randint(50, 200, (6, 30))
print(f"* Sprzedaż 6 produktów w 30 dni:\n{sales}\n")

trend = sales[:, -7:] - sales[:, :7]
print(f"* Trend sprzedaży per produkt:\n{trend}\n")

weekly_sales = [sales[:, i:i+7].sum() for i in range(0, 30, 7)]
best_week = np.argmax(weekly_sales)
print(f"* Najlepszy tydzień: {best_week + 1} (suma: {weekly_sales[best_week]})\n")

plt.figure(figsize=(12, 6))
for i in range(6):
    plt.plot(sales[i], label=f"Produkt {i+1}")
plt.xlabel("Dzień")
plt.ylabel("Sprzedaż")
plt.title("Sprzedaż per produkt w czasie")
plt.legend()
plt.show()


```
